```python
# prompt: Faça uma análise separada da loja, loja2, loja3, loja4 e adicione uma linha reta com a média de faturamento, nos 4 gráficos gerados.

import pandas as pd
import matplotlib.pyplot as plt

# ... (código anterior permanece inalterado)

# Função para plotar o gráfico de uma loja com a média
def plot_faturamento(loja, nome_loja):
    loja['Data da Compra'] = pd.to_datetime(loja['Data da Compra'], format='%d/%m/%Y')
    loja['Ano'] = loja['Data da Compra'].dt.year
    faturamento = loja.groupby(['Ano'])['Preço'].sum() - loja.groupby(['Ano'])['Frete'].sum()

    plt.figure(figsize=(10, 6))
    plt.plot(faturamento.index, faturamento.values)

    media_faturamento = faturamento.mean()
    plt.axhline(y=media_faturamento, color='r', linestyle='-', label=f'Média: {media_faturamento:.2f}')

    plt.xlabel('Ano')
    plt.ylabel('Faturamento')
    plt.title(f'Faturamento Anual - {nome_loja}')
    plt.legend()
    plt.show()

# Plotar os gráficos para cada loja
plot_faturamento(loja, 'Loja 1')
plot_faturamento(loja2, 'Loja 2')
plot_faturamento(loja3, 'Loja 3')
plot_faturamento(loja4, 'Loja 4')
```

# 2. Vendas por Categoria


```python
# prompt: Faça uma lista sem nomes repetidos na coluna Categoria do Produto para cada loja, loja2, loja3, loja4. Faça um dicionário com a chave sendo a categoria e o valor sendo as quantidades em que aparecem dos anos 2020 a 2023. Exemplo: {eletronicos, 240). Faça essa análise separada para loja, loja2, loja3, loja4. Depois gere gráficos de barra separados da loja, loja2, loja3 e loja4, contendo estas informações.

import pandas as pd
import matplotlib.pyplot as plt
def analisar_categorias(df, nome_loja):
    # Converter 'Data da Compra' para datetime e extrair o ano
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], format='%d/%m/%Y')
    df['Ano'] = df['Data da Compra'].dt.year

    # Filtrar dados de 2020 a 2023
    df_filtrado = df[(df['Ano'] >= 2020) & (df['Ano'] <= 2023)]

    # Criar dicionário de categorias e quantidades
    categorias = {}
    for categoria in df_filtrado['Categoria do Produto'].unique():
        categorias[categoria] = df_filtrado[df_filtrado['Categoria do Produto'] == categoria].shape[0]

    # Criar gráfico de barras
    plt.figure(figsize=(10, 6))
    plt.bar(categorias.keys(), categorias.values())
    plt.xlabel('Categoria do Produto')
    plt.ylabel('Quantidade')
    plt.title(f'Quantidade de Produtos por Categoria - {nome_loja}')
    plt.xticks(rotation=45, ha='right')  # Rotacionar rótulos do eixo x
    plt.tight_layout()  # Ajustar layout para evitar sobreposição
    plt.show()

    return categorias


# Analisar categorias para cada loja e gerar gráficos
categorias_loja1 = analisar_categorias(loja.copy(), 'Loja 1')
categorias_loja2 = analisar_categorias(loja2.copy(), 'Loja 2')
categorias_loja3 = analisar_categorias(loja3.copy(), 'Loja 3')
categorias_loja4 = analisar_categorias(loja4.copy(), 'Loja 4')

print("Loja 1:", categorias_loja1)
print("Loja 2:", categorias_loja2)
print("Loja 3:", categorias_loja3)
print("Loja 4:", categorias_loja4)
```

# 4. Produtos Mais e Menos Vendidos

```python
# prompt: Gere uma lista sem repetições da coluna "Produtos", para cada base de dados, loja, loja2, loja3, loja4. Depois gere um dicionário que deve conter como chave o nome da lista anterior e como valor a quantidade de vezes que o nome aparece na base de dados. Gere um gráfico que contenha os produtos com maior e menor quantidade na lista. Gere gráficos e realize as análises para cada base de dados, loja, loja2, loja3, loja4.

import pandas as pd
import matplotlib.pyplot as plt
def analisar_produtos(df, nome_loja):
    # Converter 'Data da Compra' para datetime e extrair o ano
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], format='%d/%m/%Y')
    df['Ano'] = df['Data da Compra'].dt.year

    # Filtrar dados de 2020 a 2023
    df_filtrado = df[(df['Ano'] >= 2020) & (df['Ano'] <= 2023)]

    # Lista de produtos sem repetições
    produtos_unicos = df_filtrado['Produto'].unique()

    # Dicionário com a quantidade de cada produto
    contagem_produtos = {}
    for produto in produtos_unicos:
        contagem_produtos[produto] = df_filtrado[df_filtrado['Produto'] == produto].shape[0]

    # Ordenar o dicionário pela quantidade de produtos
    produtos_ordenados = dict(sorted(contagem_produtos.items(), key=lambda item: item[1]))

    # Gerar gráfico de barras
    plt.figure(figsize=(10,6))
    plt.bar(produtos_ordenados.keys(), produtos_ordenados.values())
    plt.xlabel("Produtos")
    plt.ylabel("Quantidade Vendida")
    plt.title(f"Produtos Mais e Menos Vendidos - {nome_loja}")
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

    # Imprimir os 5 produtos mais e menos vendidos
    print(f"--- {nome_loja} ---")
    print("5 Produtos Mais Vendidos:")
    for produto, quantidade in list(produtos_ordenados.items())[-5:]:
        print(f"{produto}: {quantidade}")
    print("\n5 Produtos Menos Vendidos:")
    for produto, quantidade in list(produtos_ordenados.items())[:5]:
        print(f"{produto}: {quantidade}")

    return produtos_ordenados


# Analisar produtos para cada loja e gerar gráficos
produtos_loja1 = analisar_produtos(loja.copy(), 'Loja 1')
produtos_loja2 = analisar_produtos(loja2.copy(), 'Loja 2')
produtos_loja3 = analisar_produtos(loja3.copy(), 'Loja 3')
produtos_loja4 = analisar_produtos(loja4.copy(), 'Loja 4')
```

# 5. Frete Médio por Loja


```python
# prompt: Faça uma análise da média do valor da coluna "Frete"  nos anos de 2020 a 2023, e liste estas médias. Realize a análise nas bases de dados loja, loja2, loja3 e loja4.

import pandas as pd
def analisar_frete(df, nome_loja):
    # Converter 'Data da Compra' para datetime e extrair o ano
    df['Data da Compra'] = pd.to_datetime(df['Data da Compra'], format='%d/%m/%Y')
    df['Ano'] = df['Data da Compra'].dt.year

    # Filtrar dados de 2020 a 2023
    df_filtrado = df[(df['Ano'] >= 2020) & (df['Ano'] <= 2023)]

    # Calcular a média do frete para cada ano
    media_frete_anual = df_filtrado.groupby('Ano')['Frete'].mean()

    # Imprimir as médias
    print(f"--- {nome_loja} ---")
    print(media_frete_anual)
    print("-" * 20)

# Analisar o frete para cada loja
analisar_frete(loja.copy(), 'Loja 1')
analisar_frete(loja2.copy(), 'Loja 2')
analisar_frete(loja3.copy(), 'Loja 3')
analisar_frete(loja4.copy(), 'Loja 4')
```
